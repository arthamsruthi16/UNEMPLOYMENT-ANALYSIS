```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path=r"C:\Excel\unemployment.csv"
df = pd.read_csv(file_path)

# Clean column names
df.columns = df.columns.str.strip()

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Basic info
print(df.head())
print(df.info())



```

```python
# Drop missing values
df = df.dropna()

# Unemployment Rate Over Time
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Estimated Unemployment Rate (%)', hue='Area')
plt.title('Unemployment Rate Over Time (Rural vs Urban)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(title='Area')
plt.show()

```

```python
# State-wise Average Unemployment
state_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,8))
sns.barplot(x=state_avg.values, y=state_avg.index, palette='viridis')
plt.title('Average Unemployment Rate by State/Region')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()

```

```python
# Labour Participation vs Unemployment
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Estimated Labour Participation Rate (%)', 
                y='Estimated Unemployment Rate (%)', hue='Area')
plt.title('Labour Participation vs Unemployment Rate')
plt.xlabel('Labour Participation Rate (%)')
plt.ylabel('Unemployment Rate (%)')
plt.show()


```

```python
# Heatmap for Correlation
plt.figure(figsize=(6,4))
sns.heatmap(df[['Estimated Unemployment Rate (%)','Estimated Employed','Estimated Labour Participation Rate (%)']].corr(),
            annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
